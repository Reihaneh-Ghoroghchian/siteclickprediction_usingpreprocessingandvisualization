part 6.

This is the first method used, but this method does not correctly utilize vectorization

```python
start=time.time()
clicked_mean_age = ad_df_without_NAN[ad_df_without_NAN['Clicked on Ad']==1].mean()['Age']
not_clicked_mean_age = ad_df_without_NAN[ad_df_without_NAN['Clicked on Ad']==0].mean()['Age']
end=time.time()
print('mean age of people who clicked', clicked_mean_age)
print('mean age of people who did not click', not_clicked_mean_age)
print('time without for :',end-start)
```

Below you can see the second method in which vectorization has been used correctly, hence, the needed time for the operation has become visibly less.

```python
start=time.time()
clicked_mean_age = ((ad_df_without_NAN['Clicked on Ad']*ad_df_without_NAN['Age']).replace(0, np.NaN)).mean()
not_clicked_mean_age = ((ad_df_without_NAN['Clicked on Ad'].replace(1, np.NaN).replace(0, 1)*ad_df_without_NAN['Age']).replace(0, np.NaN)).mean()
end=time.time()
print('mean age of people who clicked', clicked_mean_age)
print('mean age of people who did not click', not_clicked_mean_age)
print('time without for :',end-start)
```

part 7.

```python
start=time.time()
total_count = ad_df_without_NAN.count()[0]
clicked_count = 0
not_clicked_count = 0
clicked_tot_age = 0
not_clicked_tot_age = 0
for i in ad_df_without_NAN.index :
    if ad_df_without_NAN['Clicked on Ad'][i]==1.0 :
        clicked_tot_age += ad_df_without_NAN['Age'][i]
        clicked_count += 1
    else:
        not_clicked_tot_age += ad_df_without_NAN['Age'][i]
        not_clicked_count += 1
end=time.time()
print('mean age in clicked', clicked_tot_age/clicked_count)
print('mean age in not clicked', not_clicked_tot_age/not_clicked_count)
print('time using for',end-start)
```

Comaparing out puts, we can see that using vectorization in part 6. decreased the time enormously. Using 'for' for big datas that are used in AI is clearly not efficient.
